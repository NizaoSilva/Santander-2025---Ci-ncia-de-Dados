```python
import pandas as pd
from datetime import date

# =========================
# 1. ETL_Extract
# =========================
# Criação do DataFrame a partir de um dicionário
dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "data_nascimento": [
        "2010-05-12",
        "2000-08-20",
        "2012-01-30",
        "1995-11-02",
        "2008-07-18"
    ]
}

df = pd.DataFrame(dados)

# =========================
# 2. ETL_Transform
# =========================
# Converte para datetime
df["data_nascimento"] = pd.to_datetime(df["data_nascimento"])

# Calcula a idade
hoje = pd.Timestamp(date.today())
df["idade"] = (hoje - df["data_nascimento"]).dt.days // 365

# =========================
# 3. ETL_Load
# =========================
# Filtra pessoas com menos de 18 anos
menores_18 = df[df["idade"] < 18]

# Exibe o resultado
print("Pessoas com menos de 18 anos:")
print(menores_18)
```
